Convert the image to grayscale

```python
import cv2
image = cv2.imread('exit-ramp.png')
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)
kernel_size = 3
blur_gray = cv2.GaussianBlur(gray, (kernel_size, kernel_size),0)
%matplotlib inline
low_threshold = 100
high_threshold = 300
edges1 = cv2.Canny(blur_gray, low_threshold, high_threshold)
edges2 = cv2.Canny(blur_gray, 50, 240)

fig = plt.figure()http://docs.opencv.org/3.0-beta/_images/houghlines1.svg
a = fig.add_subplot(1,2,1)
a.set_title('Before')
edge1plot = plt.imshow(edges1, cmap="Greys_r")

a = fig.add_subplot(1, 2, 2)
edge1plot = plt.imshow(edges2, cmap="Greys_r")
a.set_title('After')

plt.show()
```
